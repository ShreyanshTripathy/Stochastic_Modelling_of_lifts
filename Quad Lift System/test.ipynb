```python
import random

# ðŸ›— **Lift Initialization**
lifts = {
    f"Lift_{i+1}": {
        "current_floor": [1, 1, 5, 5][i],
        "direction": 0,  # 1 for up, -1 for down, 0 for idle
        "passengers": [],
        "pending_orders": [],
        "status": [False, False, False, False][i]  # False for idle, True for moving
    }
    for i in range(4)
}

# ðŸŽŸï¸ **Pending Orders**
pending_orders = [
    (1, 0, 17, 0, 0, 0, 1), (2, 1, 19, 0, 0, 0, 1), (3, 2, 20, 0, 0, 0, 1),
    (4, 3, 10, 0, 0, 0, 1), (5, 4, 7, 0, 0, 0, 1), (6, 5, 11, 0, 0, 0, 1),
    (7, 6, 11, 0, 0, 0, 1), (8, 7, 2, 0, 0, 0, -1), (9, 8, 12, 0, 0, 0, 1),
    (10, 9, 12, 0, 0, 0, 1), (11, 10, 13, 0, 0, 0, 1), (12, 11, 4, 0, 0, 0, -1),
    (13, 12, 19, 0, 0, 0, 1)
]

# ðŸ“š **Helper Functions**

def calculate_lift_distances(passenger_floor):
    """Calculate distances of passenger from each lift."""
    return {
        lift_name: abs(lift["current_floor"] - passenger_floor)
        for lift_name, lift in lifts.items()
    }

def get_nearest_lifts(distances):
    """Get all lifts at the minimum distance."""
    min_distance = min(distances.values())
    return [lift for lift, distance in distances.items() if distance == min_distance]

def is_lift_available(lift, direction, passenger_floor):
    """Check if the lift is available for the passenger."""
    return (
        not lift["status"] or
        (lift["direction"] == 0 or
         (lift["direction"] == direction and
          ((direction == 1 and passenger_floor >= lift["current_floor"]) or
           (direction == -1 and passenger_floor <= lift["current_floor"]))))
    )

def assign_passenger_to_lift(lift_name, passenger):
    """Assign a passenger to the specified lift."""
    lift = lifts[lift_name]
    lift["pending_orders"].append(passenger)
    lift["status"] = True

    _, passenger_floor, _, _, _, _, direction = passenger
    if lift["direction"] == 0:
        lift["direction"] = (
            1 if passenger_floor > lift["current_floor"] else
            -1 if passenger_floor < lift["current_floor"] else
            direction
        )

def handle_idle_lift(lift_name, passenger):
    """Assign passenger to an idle lift and set its direction."""
    lift = lifts[lift_name]
    _, passenger_floor, _, _, _, _, direction = passenger
    assign_passenger_to_lift(lift_name, passenger)
    lift["direction"] = (
        1 if passenger_floor > lift["current_floor"] else
        -1 if passenger_floor < lift["current_floor"] else
        direction
    )

# ðŸš€ **Same-Floor Optimization**
def handle_same_floor_passenger(passenger):
    """Check if a lift is on the same floor and moving in the passenger's direction."""
    _, passenger_floor, _, _, _, _, direction = passenger

    for lift_name, lift in lifts.items():
        if (
            lift["current_floor"] == passenger_floor and  # Same floor
            lift["direction"] == direction and  # Same direction
            passenger not in lift["pending_orders"]  # Avoid duplicate assignment
        ):
            assign_passenger_to_lift(lift_name, passenger)
            return lift_name  # Return the lift assigned
    return None  # No same-floor lift available

def assign_passengers(pending_orders):
    """Main function to assign passengers to lifts."""
    assignments = {lift_name: [] for lift_name in lifts}

    for passenger in pending_orders:
        _, passenger_floor, _, _, _, _, direction = passenger

        # Step 1: Same-Floor Optimization
        same_floor_lift = handle_same_floor_passenger(passenger)
        if same_floor_lift:
            assignments[same_floor_lift].append(passenger)
            continue

        # Step 2: Find nearest lifts
        distances = calculate_lift_distances(passenger_floor)
        nearest_lifts = get_nearest_lifts(distances)

        # Step 3: Assign to the most suitable lift
        for lift_name in nearest_lifts:
            if is_lift_available(lifts[lift_name], direction, passenger_floor):
                assign_passenger_to_lift(lift_name, passenger)
                assignments[lift_name].append(passenger)
                break
        else:
            # Step 4: Assign to an idle lift if no suitable lift found
            idle_lifts = [lift_name for lift_name, lift in lifts.items() if not lift["status"]]
            if idle_lifts:
                handle_idle_lift(idle_lifts[0], passenger)
                assignments[idle_lifts[0]].append(passenger)
            else:
                # Step 5: Random fallback assignment
                random_lift = random.choice(nearest_lifts)
                assign_passenger_to_lift(random_lift, passenger)
                assignments[random_lift].append(passenger)

    return assignments

# ðŸ“Š **Execution**
if __name__ == "__main__":
    lift_assignments = assign_passengers(pending_orders)
    for lift_name, orders in lift_assignments.items():
        print(f"\n{lift_name} Assigned Orders:")
        for order in orders:
            print(order)
```
