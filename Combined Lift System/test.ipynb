```python
import numpy as np
lambda_passenger_per_floor = [0.003] + [0.004]*5
for floor in range(1):
    inter_arrival_time = np.random.exponential(1 / lambda_passenger_per_floor[floor])
    print(f"Time between passengers on floor {floor}: {inter_arrival_time:.2f} seconds")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the arrival rate (lambda) for floor 0
lambda_floor_0 = 0.003  # Passengers per second

# Generate multiple inter-arrival times
num_samples = 1000  # Number of passengers
inter_arrival_times = np.random.exponential(1 / lambda_floor_0, num_samples)

# Plot the histogram of inter-arrival times
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(inter_arrival_times, bins=30, density=True, alpha=0.7, color='blue')
plt.xlabel("Inter-Arrival Time (seconds)")
plt.ylabel("Density")
plt.title("Histogram of Passenger Inter-Arrival Times")

# Plot the first 50 inter-arrival times as a time series
plt.subplot(1, 2, 2)
plt.plot(np.cumsum(inter_arrival_times[:50]), marker='o', linestyle='-')
plt.xlabel("Passenger Number")
plt.ylabel("Time (seconds)")
plt.title("Passenger Arrival Times (First 50)")

plt.tight_layout()
plt.show()
```
